## 0) Tratamento e Limpeza do Texto

```python
from bs4 import BeautifulSoup
import nltk 
import string 
import re
from nltk.corpus import stopwords
import pandas as pd
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
ptbr_stopwords = stopwords.words('portuguese')
```

```python
ptbr_stopwords.append('vol')
```

```python
for i in range(97,123):
		ptbr_stopwords.append(chr(i))
```

```python
def remove_html(text):
    soup = BeautifulSoup(text, 'lxml')
    html_free = soup.text
    return html_free
```

```python
def remove_punctuation(text):
    clean_txt = "".join([c for c in text if c not in string.punctuation])
    return clean_txt
```

```python
def to_lower(text):
    return [w.lower() for w in text.split()]
```

```python
def remove_stopwords(text):
    phrase = [w.lower() for w in text if w not in ptbr_stopwords]
    return phrase
```

### 0.1) Le o arquivo

```python
file = open('Helena.txt', encoding='latin-1', mode='r')
```

```python
data = file.read()
```

### 0.2) Limpa o texto

```python
data = remove_html(data)
data = remove_punctuation(data)
data = to_lower(data)
data = remove_stopwords(data)
```

```python
data = pd.DataFrame(data=data,columns=['words'])
```

## 1) Quantidade de palavras distintas

```python
unique_words = data['words'].unique()
print("A quantidade de palavras distintas no texto é: ",len(unique_words))
```

## 2) Histograma das palavras

Nesta etapa foi feito histograma das 30 palavras mais frequentes no texto

```python
import collections
```

```python
d = collections.Counter(data['words'])
```

```python
most_common_df = pd.DataFrame(d.most_common(), columns=['words', 'frequency'])
```

```python
most_common_df.set_index(keys='words')[:30].plot(kind='bar', figsize=(16,5), title="Word's Histogram")
```

## 3) Histograma de prefixos

```python
prefixes = {1:{}, 
            2:{},
            3:{},
            4:{},
            5:{}
           }
```

```python
sufixes = {1:{}, 
            2:{},
            3:{},
            4:{},
            5:{}
           }
```

```python
for key in prefixes.keys():
    for w in data['words']:
        if w[:key] not in prefixes[key]:
            prefixes[key][w[:key]] = 1
        else:
            prefixes[key][w[:key]] += 1
    pd.DataFrame(prefixes[key].items(), columns=['prefix', 'amount']).sort_values('amount', ascending=False).set_index('prefix')[:50].plot(kind='bar', figsize=(16,5), title="Histogram - 50 Prefixes of Size {}".format(key))
```

## 4) Histograma de sufixos

```python
for key in sufixes.keys():
    for w in data['words']:
        if w[-key:] not in sufixes[key]:
            sufixes[key][w[-key:]] = 1
        else:
            sufixes[key][w[-key:]] += 1
    pd.DataFrame(sufixes[key].items(), columns=['prefix', 'amount']).sort_values('amount', ascending=False).set_index('prefix')[:50].plot(kind='bar', figsize=(16,5), title="Histogram - 50 Sufixes of Size {}".format(key))
```
